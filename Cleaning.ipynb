```python
import pandas as pd
import numpy as np

# 1. Helper function to get the text category (Bucket)
def get_aqi_bucket(aqi_value):
    if pd.isnull(aqi_value): return np.nan
    if aqi_value <= 50: return "Good"
    elif aqi_value <= 100: return "Satisfactory"
    elif aqi_value <= 200: return "Moderate"
    elif aqi_value <= 300: return "Poor"
    elif aqi_value <= 400: return "Very Poor"
    elif aqi_value > 400: return "Severe"
    else: return np.nan

# 2. The Main Formula to Calculate AQI
def calculate_aqi(row):
    # Indian National Air Quality Index (NAQI) Breakpoints
    breakpoints = {
        'PM2.5': [(0, 30, 0, 50), (30, 60, 51, 100), (60, 90, 101, 200), (90, 120, 201, 300), (120, 250, 301, 400), (250, 9999, 401, 500)],
        'PM10':  [(0, 50, 0, 50), (50, 100, 51, 100), (100, 250, 101, 200), (250, 350, 201, 300), (350, 430, 301, 400), (430, 9999, 401, 500)],
        'NO2':   [(0, 40, 0, 50), (40, 80, 51, 100), (80, 180, 101, 200), (180, 280, 201, 300), (280, 400, 301, 400), (400, 9999, 401, 500)],
        'SO2':   [(0, 40, 0, 50), (40, 80, 51, 100), (80, 380, 101, 200), (380, 800, 201, 300), (800, 1600, 301, 400), (1600, 9999, 401, 500)],
        'CO':    [(0, 1, 0, 50), (1, 2, 51, 100), (2, 10, 101, 200), (10, 17, 201, 300), (17, 34, 301, 400), (34, 9999, 401, 500)],
        'O3':    [(0, 50, 0, 50), (50, 100, 51, 100), (100, 168, 101, 200), (168, 208, 201, 300), (208, 748, 301, 400), (748, 9999, 401, 500)],
        'NH3':   [(0, 200, 0, 50), (200, 400, 51, 100), (400, 800, 101, 200), (800, 1200, 201, 300), (1200, 1800, 301, 400), (1800, 9999, 401, 500)]
    }
    
    aqi_list = []
    
    # Loop through each pollutant and calculate its specific sub-index
    for pollutant, ranges in breakpoints.items():
        if pollutant in row and not pd.isnull(row[pollutant]):
            concentration = row[pollutant]
            for (low, high, aqi_low, aqi_high) in ranges:
                if low <= concentration <= high:
                    # The formula: I = [(I_Hi - I_Lo) / (C_Hi - C_Lo)] * (C - C_Lo) + I_Lo
                    aqi = ((aqi_high - aqi_low) / (high - low)) * (concentration - low) + aqi_low
                    aqi_list.append(aqi)
                    break
    
    # The final AQI is the MAXIMUM of all sub-indices
    if aqi_list:
        return max(aqi_list)
    else:
        return np.nan

print("Functions defined successfully!")
```

```python
# 1. Interpolate what we can (Standard Pollutants)
# We do NOT fill with 0. We only connect the dots where dots exist.
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

for col in pollutants:
    city_day[col] = city_day.groupby('City')[col].transform(
        lambda x: x.interpolate(method='linear', limit_direction='both')
    )

print("Interpolation done. Remaining gaps are genuine missing data.")

# 2. Check remaining nulls (Likely Lucknow's PM10 and the 14 cities' Xylene)
print(city_day[['PM10', 'Xylene']].isnull().sum())

# 3. Calculate AQI (The formula ignores NaNs automatically)
# Ensure you have the 'calculate_aqi' function defined from before
print("Calculating AQI...")
mask_missing_aqi = city_day['AQI'].isnull()
city_day.loc[mask_missing_aqi, 'AQI'] = city_day.loc[mask_missing_aqi].apply(calculate_aqi, axis=1)

# 4. Fill Bucket
city_day['AQI_Bucket'] = city_day['AQI'].apply(get_aqi_bucket)

print("Process Complete. No fake zeros added!")
```

```python
import pandas as pd
import numpy as np

print("--- STARTING FRESH REPAIR ---")

# 1. LOAD THE DATA FROM SCRATCH (Wipes out all previous mistakes)
# (Make sure the filenames match what you have on your system)
station_day = pd.read_csv('station_day.csv')
stations = pd.read_csv('stations.csv')

# 2. MERGE CITY INFO (We need this to know which city a station belongs to)
station_day['Date'] = pd.to_datetime(station_day['Date'])
station_day = station_day.merge(stations[['StationId', 'City']], on='StationId', how='left')

# 3. DEFINE THE FORMULAS (The Recipes)
def get_PM25_subindex(x):
    if x <= 30: return x * 50 / 30
    elif x <= 60: return 50 + (x - 30) * 50 / 30
    elif x <= 90: return 100 + (x - 60) * 100 / 30
    elif x <= 120: return 200 + (x - 90) * 100 / 30
    elif x <= 250: return 300 + (x - 120) * 100 / 130
    elif x > 250: return 400 + (x - 250) * 100 / 130
    else: return 0

def get_PM10_subindex(x):
    if x <= 50: return x
    elif x <= 100: return x
    elif x <= 250: return 100 + (x - 100) * 100 / 150
    elif x <= 350: return 200 + (x - 250) * 100 / 100
    elif x <= 430: return 300 + (x - 350) * 100 / 80
    elif x > 430: return 400 + (x - 430) * 100 / 80
    else: return 0

def get_SO2_subindex(x):
    if x <= 40: return x * 50 / 40
    elif x <= 80: return 50 + (x - 40) * 50 / 40
    elif x <= 380: return 100 + (x - 80) * 100 / 300
    elif x <= 800: return 200 + (x - 380) * 100 / 420
    elif x <= 1600: return 300 + (x - 800) * 100 / 800
    elif x > 1600: return 400 + (x - 1600) * 100 / 800
    else: return 0

def get_NOx_subindex(x):
    if x <= 40: return x * 50 / 40
    elif x <= 80: return 50 + (x - 40) * 50 / 40
    elif x <= 180: return 100 + (x - 80) * 100 / 100
    elif x <= 280: return 200 + (x - 180) * 100 / 100
    elif x <= 400: return 300 + (x - 280) * 100 / 120
    elif x > 400: return 400 + (x - 400) * 100 / 120
    else: return 0

def get_NH3_subindex(x):
    if x <= 200: return x * 50 / 200
    elif x <= 400: return 50 + (x - 200) * 50 / 200
    elif x <= 800: return 100 + (x - 400) * 100 / 400
    elif x <= 1200: return 200 + (x - 800) * 100 / 400
    elif x <= 1800: return 300 + (x - 1200) * 100 / 600
    elif x > 1800: return 400 + (x - 1800) * 100 / 600
    else: return 0

def get_CO_subindex(x):
    if x <= 1: return x * 50 / 1
    elif x <= 2: return 50 + (x - 1) * 50 / 1
    elif x <= 10: return 100 + (x - 2) * 100 / 8
    elif x <= 17: return 200 + (x - 10) * 100 / 7
    elif x <= 34: return 300 + (x - 17) * 100 / 17
    elif x > 34: return 400 + (x - 34) * 100 / 17
    else: return 0

def get_O3_subindex(x):
    if x <= 50: return x * 50 / 50
    elif x <= 100: return 50 + (x - 50) * 50 / 50
    elif x <= 168: return 100 + (x - 100) * 100 / 68
    elif x <= 208: return 200 + (x - 168) * 100 / 40
    elif x <= 748: return 300 + (x - 208) * 100 / 539
    elif x > 748: return 400 + (x - 748) * 100 / 539
    else: return 0

def get_aqi_bucket(x):
    if pd.isnull(x): return 'Unknown'
    if x <= 50: return 'Good'
    elif x <= 100: return 'Satisfactory'
    elif x <= 200: return 'Moderate'
    elif x <= 300: return 'Poor'
    elif x <= 400: return 'Very Poor'
    else: return 'Severe'

# 4. THE CLEANUP (The Smart Strategy)
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3']

print("1. Filling missing pollutant data...")
# A. First, try to fill using the same station (interpolation)
# This is good for small gaps (missing 1 day)
for col in pollutants:
    station_day[col] = station_day.groupby('StationId')[col].transform(
        lambda x: x.interpolate(method='linear', limit_direction='both')
    )

# B. Second, for big gaps (empty station), fill using the CITY Average for that day
for col in pollutants:
    station_day[col] = station_day[col].fillna(
        station_day.groupby(['City', 'Date'])[col].transform('median')
    )

# C. Third, fallback to Country Average if City is also empty
for col in pollutants:
    station_day[col] = station_day[col].fillna(
        station_day.groupby('Date')[col].transform('median')
    )

# D. Final safety net (Forward/Back fill)
station_day[pollutants] = station_day[pollutants].ffill().bfill()
```

```python
import pandas as pd
import numpy as np

print("--- PROCESSING STATION_HOUR (This may take a minute) ---")

# 1. SETUP (Merge City Names)
# We assume station_hour is already loaded. If not, uncomment the next line:
# station_hour = pd.read_csv('station_hour.csv')
station_hour['Datetime'] = pd.to_datetime(station_hour['Datetime'])

if 'City' not in station_hour.columns:
    # Ensure 'stations' dataframe exists (from previous steps)
    station_hour = station_hour.merge(stations[['StationId', 'City']], on='StationId', how='left')

# 2. FILL MISSING POLLUTANTS (The Smart Strategy)
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3']

print("1. Filling missing pollutant data...")

# A. Interpolate within the same station (Good for short gaps of a few hours)
# We limit direction to avoid drawing lines across massive month-long gaps
for col in pollutants:
    station_hour[col] = station_hour.groupby('StationId')[col].transform(
        lambda x: x.interpolate(method='linear', limit_direction='both')
    )

# B. Fill remaining gaps using the CITY Median for that specific hour
# (If Station A is dead at 10 AM, use the average of Station B & C at 10 AM)
print("   ...Applying City-wide averages (Heavy Step)...")
for col in pollutants:
    station_hour[col] = station_hour[col].fillna(
        station_hour.groupby(['City', 'Datetime'])[col].transform('median')
    )

# C. Fallback: Global Median for that hour (if the whole city is offline)
for col in pollutants:
    station_hour[col] = station_hour[col].fillna(
        station_hour.groupby('Datetime')[col].transform('median')
    )

# D. Final Safety Net (Forward/Back fill)
station_hour[pollutants] = station_hour[pollutants].ffill().bfill()

# 3. RE-DEFINE FORMULAS (Safety measure)
def get_PM25_subindex(x):
    if x <= 30: return x * 50 / 30
    elif x <= 60: return 50 + (x - 30) * 50 / 30
    elif x <= 90: return 100 + (x - 60) * 100 / 30
    elif x <= 120: return 200 + (x - 90) * 100 / 30
    elif x <= 250: return 300 + (x - 120) * 100 / 130
    elif x > 250: return 400 + (x - 250) * 100 / 130
    else: return 0

def get_PM10_subindex(x):
    if x <= 50: return x
    elif x <= 100: return x
    elif x <= 250: return 100 + (x - 100) * 100 / 150
    elif x <= 350: return 200 + (x - 250) * 100 / 100
    elif x <= 430: return 300 + (x - 350) * 100 / 80
    elif x > 430: return 400 + (x - 430) * 100 / 80
    else: return 0

def get_SO2_subindex(x):
    if x <= 40: return x * 50 / 40
    elif x <= 80: return 50 + (x - 40) * 50 / 40
    elif x <= 380: return 100 + (x - 80) * 100 / 300
    elif x <= 800: return 200 + (x - 380) * 100 / 420
    elif x <= 1600: return 300 + (x - 800) * 100 / 800
    elif x > 1600: return 400 + (x - 1600) * 100 / 800
    else: return 0

def get_NOx_subindex(x):
    if x <= 40: return x * 50 / 40
    elif x <= 80: return 50 + (x - 40) * 50 / 40
    elif x <= 180: return 100 + (x - 80) * 100 / 100
    elif x <= 280: return 200 + (x - 180) * 100 / 100
    elif x <= 400: return 300 + (x - 280) * 100 / 120
    elif x > 400: return 400 + (x - 400) * 100 / 120
    else: return 0

def get_NH3_subindex(x):
    if x <= 200: return x * 50 / 200
    elif x <= 400: return 50 + (x - 200) * 50 / 200
    elif x <= 800: return 100 + (x - 400) * 100 / 400
    elif x <= 1200: return 200 + (x - 800) * 100 / 400
    elif x <= 1800: return 300 + (x - 1200) * 100 / 600
    elif x > 1800: return 400 + (x - 1800) * 100 / 600
    else: return 0

def get_CO_subindex(x):
    if x <= 1: return x * 50 / 1
    elif x <= 2: return 50 + (x - 1) * 50 / 1
    elif x <= 10: return 100 + (x - 2) * 100 / 8
    elif x <= 17: return 200 + (x - 10) * 100 / 7
    elif x <= 34: return 300 + (x - 17) * 100 / 17
    elif x > 34: return 400 + (x - 34) * 100 / 17
    else: return 0

def get_O3_subindex(x):
    if x <= 50: return x * 50 / 50
    elif x <= 100: return 50 + (x - 50) * 50 / 50
    elif x <= 168: return 100 + (x - 100) * 100 / 68
    elif x <= 208: return 200 + (x - 168) * 100 / 40
    elif x <= 748: return 300 + (x - 208) * 100 / 539
    elif x > 748: return 400 + (x - 748) * 100 / 539
    else: return 0

def get_aqi_bucket(x):
    if pd.isnull(x): return 'Unknown'
    if x <= 50: return 'Good'
    elif x <= 100: return 'Satisfactory'
    elif x <= 200: return 'Moderate'
    elif x <= 300: return 'Poor'
    elif x <= 400: return 'Very Poor'
    else: return 'Severe'

# 4. CALCULATE AQI
print("2. Calculating AQI (This is the slow part)...")

station_hour['PM2.5_SubIndex'] = station_hour['PM2.5'].apply(lambda x: get_PM25_subindex(x))
station_hour['PM10_SubIndex'] = station_hour['PM10'].apply(lambda x: get_PM10_subindex(x))
station_hour['SO2_SubIndex'] = station_hour['SO2'].apply(lambda x: get_SO2_subindex(x))
station_hour['NOx_SubIndex'] = station_hour['NOx'].apply(lambda x: get_NOx_subindex(x))
station_hour['NH3_SubIndex'] = station_hour['NH3'].apply(lambda x: get_NH3_subindex(x))
station_hour['CO_SubIndex'] = station_hour['CO'].apply(lambda x: get_CO_subindex(x))
station_hour['O3_SubIndex'] = station_hour['O3'].apply(lambda x: get_O3_subindex(x))

station_hour['Calculated_AQI'] = station_hour[[
    'PM2.5_SubIndex', 'PM10_SubIndex', 'SO2_SubIndex', 
    'NOx_SubIndex', 'NH3_SubIndex', 'CO_SubIndex', 'O3_SubIndex'
]].max(axis=1)

# Overwrite AQI
station_hour['AQI'] = station_hour['Calculated_AQI']
station_hour['AQI_Bucket'] = station_hour['AQI'].apply(get_aqi_bucket)

# 5. CLEANUP
cols_to_drop = [
    'PM2.5_SubIndex', 'PM10_SubIndex', 'SO2_SubIndex', 
    'NOx_SubIndex', 'NH3_SubIndex', 'CO_SubIndex', 'O3_SubIndex', 
    'Calculated_AQI'
]
station_hour.drop(columns=cols_to_drop, inplace=True, errors='ignore')

print("--- STATION_HOUR COMPLETE ---")
print("Remaining Missing AQI:", station_hour['AQI'].isnull().sum())
print(station_hour[['StationId', 'Datetime', 'AQI', 'AQI_Bucket']].tail())
```
